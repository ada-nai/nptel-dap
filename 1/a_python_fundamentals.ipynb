# Indexing in DFs

## `loc`

```python
# Access first row (zero-indexing)
df.loc[0]
```

```python
# Access 100th row 
df.loc[99]
```

```python
# Select 1st, 100th, 1000th row
df.loc[[0, 99, 999]]
```

## `iloc`
Indexed location

```python
# Access second record
df.iloc[1]
```

```python
# Negative indexing (last record)
df.iloc[-1]
```

# Subsetting columns

## Subset using `loc`

```python
# Obtain all records for only two columns ['year', 'pop']
subset = df.loc[:, ['year', 'pop']]
subset.head()
```

## Subset using `iloc`


```python
subset = df.iloc[:, [2, 4, -1]]
subset.head()
```

```python
# Subset using range
small_range = range(6)
df.iloc[:5, small_range]
```

# Subsetting rows and columns

```python
df.loc[42, 'country']
```

```python
# Get 1st, 100th, 1000th rows from 1st, 4th, 6th column
df.iloc[[0, 99, 999], [0, 3, 5]]
```

```python
df.loc[[0, 99, 999], ['country', 'lifeExp', 'gdpPercap']]
```

# Aggregation using `groupby`

```python
# Obtain mean average life expectancy per country
df.groupby('year')['lifeExp'].mean()
```

```python
# multi groupby
multi = df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean()
multi.head(10)
```

# Flatten DF

```python
flat = multi.reset_index()
flat.head()
```

# Get unique values 

```python
df.groupby('continent')['country'].nunique()
```

# Basic plot

```python
yearly_life_expectancy = df.groupby('year')['lifeExp'].mean()
yearly_life_expectancy.head()
```

```python
plt.plot(yearly_life_expectancy)
plt.title('Year wise mean global life expectancy')
```
